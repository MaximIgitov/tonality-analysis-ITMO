```python
#Импорт библиотек
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
#записываем метод векторизации в переменную vectorizer
vectorizer = TfidfVectorizer()
```

```python
#Чтение файла из предыдушей ЛР
df = pd.read_csv(r'/content/_Manual_transcription_with_normalaize (1).csv')
df.head()
```

```python
vector_df = pd.DataFrame(0, index = range(len(df)), columns = range(3127))
vector_df.head()
```

```python
#Добавление значений вектора в строки df
for i in range(len(df)):
  x = vectorizer.fit_transform((df.at[i, 'Normalize transcription']).split()).toarray()
  st = list(x.ravel())
  for j in range(len(st)):
    vector_df[i][j] = st[j]
print(vector_df)
```

```python
#Добавление столбца "Sentiment"
vector_df['Sentiment'] = df['Sentiment digits']
vector_df.tail()
```

```python
#Сохранение файла
vector_df.to_csv(r'\TF-IDF_vector.csv', index=False)
```
