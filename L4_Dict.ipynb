```python
# Импорт библиотек
import pandas as pd
import numpy as np
import nltk
import pymorphy2
nltk.download('punkt')
nltk.download('stopwords')
from nltk.tokenize import word_tokenize
from string import punctuation
punctuation = list(punctuation)
from nltk.stem.snowball import RussianStemmer
stemmer = RussianStemmer()
```

```python
def normalize_text(s):
    # Токенизация
    tokens = word_tokenize(s)

    # Удаляем пунктуацию
    tokens_without_punct = [i for i in tokens if i not in punctuation]
    low_tokens = [i.lower() for i in tokens_without_punct]

    # удаляем стоп-слова из нашего текста
    stopwords = nltk.corpus.stopwords.words('russian')
    words_without_stop = [i for i in low_tokens if i not in stopwords]

    # Лемматизация
    morph = pymorphy2.MorphAnalyzer()
    lemms = [morph.parse(i)[0].normal_form for i in words_without_stop]

    # Стемминг
    #stems = [stemmer.stem(i) for i in lemms]

    # Вывод значения в строке
    total=''
    for el in lemms:
      total+=el
      total+=' '
    return total
```

```python
# Ввод данных
def dict_vect(sentence):
    norm_text = normalize_text(sentence).split()

    vector_df = pd.DataFrame({'num':[]})
    vector_df['num'] = np.zeros(len(norm_text))

    vector = []

    # Создание вектора
    for i in range(len(norm_text)):
        if norm_text[i] in dictionary.values:
            vector_df.at[i, 'num'] = dictionary[dictionary['Word'] == norm_text[i]]['Weight']
        else:
            vector_df.at[i, 'num'] = 0
    for j in range(len(vector_df)):
        vector.append(vector_df.at[j, 'num'])
    vector = list(vector)
    str_vector = ''
    for i in range(len(vector)):
        str_vector += (str(vector[i])+" ")
    #Вывод вектора и его длинны
    return str_vector

df['Vector'] = np.zeros(len(df))

for i in range(len(df)):
    df.at[i, 'Vector'] = dict_vect(df.at[i, 'Normalize transcription'])

df.head()



```

```python
#Подсчет кол-ва отрицательных и положительных слов
def static(s):
    vector = s.split()
    vector = [int(i) for i in range(len(vector))]
    num_pos, num_null, num_neg = 0, 0, 0
    for el in vector:
        if el>0:
            num_pos+=1
        elif el < 0:
            num_neg+=1
        else:
            num_null+=1

    # Внесение в стат. вектор
    static_vector = []
    static_vector.append(num_pos)
    static_vector.append(num_neg)

    #Среднее значение сентимента
    static_vector.append(sum(vector)/len(vector))

    #Минимум и максимум
    static_vector.append(max(vector))
    static_vector.append(min(vector))

    str_static_vector = ''
    for i in range(len(static_vector)):
        str_static_vector += (str(static_vector[i])+" ")
    return str_static_vector

```

```python
df['Static vector'] = np.zeros(len(df))
for i in range(len(df)):
    df.at[i, 'Static vector'] = static(df.at[i, 'Vector'])

df.head()
```
